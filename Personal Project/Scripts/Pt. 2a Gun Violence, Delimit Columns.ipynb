## 02. Delimit 'gun_type' Column

```python
#create df for just gun_type

df_type = df[['incident_id', 'gun_type']]
df_type.head()
```

```python
#delimit 'gun_type'

df_type = df_type['gun_type'].apply(lambda x: [item.split('::')[1] if isinstance(x, str) 
                                                            and '::' in item else [np.nan] for item in x.split('||')] 
                                                 if isinstance(x, str) else [np.nan])
df_type = pd.DataFrame(df_type.tolist())
```

```python
df_type
```

```python
df_type[0].value_counts(dropna = False)
```

```python
#replace unknown, none, and other with NaN

df_type = df_type.replace(['Unknown', 'Other', 'None'], np.nan)
```

```python
#check value counts

df_type[0].value_counts(dropna = False)
```

```python
df_type.shape
```

```python
#add incident_id 
  
df_type.insert(0, 'incident_id', df['incident_id'])
df_type.head()
```
