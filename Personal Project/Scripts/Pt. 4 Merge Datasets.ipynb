## 02. Prepare and Perform Merges



```python
#rearrange cols so state is followed by year

df_inc = df_inc[['incident_id', 'state', 'year', 'n_killed', 'n_injured', 'n_guns_involved', 'latitude', 'longitude']]
df_pol = df_pol[['state', 'year', 'state_control']]
df_pop = df_pop[['state', 'year', 'population']]
df_pop.head()
```

```python
#merge laws and legislative party dfs

df_pol_laws = pd.merge(df_pol, df_laws, on = ['state', 'year'])
df_pol_laws.head(10)
```

## 03. Export  Data Sets

```python
df_pol_laws.to_csv(os.path.join(path, 'Data', 'Final Data', 'pol_laws.csv'), index = False)
df_pop.to_csv(os.path.join(path, 'Data', 'Intermediary Data', 'pop_ordered.csv'), index = False)
df_inc.to_csv(os.path.join(path, 'Data', 'Final Data', 'inc_ordered.csv'), index = False)
df_pol.to_csv(os.path.join(path, 'Data', 'Intermediary Data', 'pol_ordered.csv'), index = False)
```
