## 04. Import Population Data

## 05. Clean and Transform Population Data

```python
#remove 2013 and 2018

df_pop.drop(columns = ['2013', '2018'], inplace = True)
df_pop.head()
```

```python
#change state col to lowercase

df_pop.rename(columns = {'State' : 'state'}, inplace = True)
```

```python
#transform df to different format

df_pop = df_pop.set_index('state').unstack().reset_index().rename(columns={'level_0':'year',0:'population'})
df_pop.head()
```
