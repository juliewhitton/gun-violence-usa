```python
#transform df_leg to match other dfs

df_leg = df_leg.set_index('state').unstack().reset_index().rename(columns={'level_0':'year',0:'state_control'})
df_leg.head()
```

```python
#add DC info that had been excluded
dc_list = [pd.Series(['2014', 'DC', 'Dem'], index = df_leg.columns), 
           pd.Series(['2015', 'DC', 'Dem'], index = df_leg.columns),
           pd.Series(['2016', 'DC', 'Dem'], index = df_leg.columns),
           pd.Series(['2017', 'DC', 'Dem'], index = df_leg.columns)]

df_leg = df_leg.append(dc_list, ignore_index = True)
```
